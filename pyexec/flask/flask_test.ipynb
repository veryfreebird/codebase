```python
import random
import pathlib
from flask import Flask, Response

app = Flask(__name__)
app.config['JSON_AS_ASCII'] = False

def get_bus_info(station):
    info = []
    for random_bus in range(10):
        bus_name = f"W{random.randint(1, 100)}"
        bus_arrival_time = random.randint(1, 30)
        bus_info = f"{bus_name} 还有 {bus_arrival_time} 分钟到达 {station}"
        info.append({bus_name: bus_info})
    return info

def discover_reports():
    """查找所有的测试报告"""
    import pathlib
    report_dir = pathlib.Path('.').parent / 'output'
    reports = [f.name for f in report_dir.iterdir() if f.suffix == '.html']
    return reports


@app.route('/')
def index():
    return {
        "msg" : "success",
        "data": {
            "id" : 'default one',
            "username" : 'test',
            "age" : 18
        }
    }


@app.route('/user/<u_id>')
def user_info(u_id):
    return {
        "msg" : "success",
        "data": {
            "id" : u_id,
            "username" : 'test',
            "age" : 18
        }
    }

@app.route('/bus/<station>')
def bus_info(station):
    info = get_bus_info(station)
    return {
        "msg": "success",
        "data": info
    }

@app.route('/report/<file_name>')
def get_report(file_name):
    """根据文件名获取测试报告。"""
    file_path = pathlib.Path('.').parent / 'output' / file_name
    f = open(file_path)##, encoding='utf-8')
    report = f.read()
    f.close()
    return Response(report)

@app.route('/reports')
def get_reports():
    report_names = discover_reports()
    return {
        "msg": "success",
        "data": [f"http://localhost:5000/report/{name}"
                 for name in report_names
                 ]
    }

@app.route('/reportsv1')
def get_reportsv1():
    report_names = discover_reports()
    data = [f"http://localhost:5000/report/{name}"
                 for name in report_names
                 ]
    return Response(data,200)

app.run()
```
